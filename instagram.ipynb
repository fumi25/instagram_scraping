```python
from selenium import webdriver
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
USERNAME = ""
PASSWORD = ""
```

```python
chrome_driver_path = r"C:\chrom_driver\chromedriver-win64\chromedriver-win64\chromedriver.exe"

service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service)
```

```python
target_url = "https://www.instagram.com/"
driver.get(target_url)
sleep(3)
```

```python
error_flg = False
```

```python
try:
    username_input = driver.find_element(By.XPATH, "//input[@aria-label='電話番号、ユーザーネーム、またはメールアドレス']")
    username_input.send_keys(USERNAME)
    sleep(1)

    password_input = driver.find_element(By.XPATH, "//input[@aria-label='パスワード']")
    password_input.send_keys(PASSWORD)
    sleep(1)

    login_button = driver.find_element(By.XPATH, "//button[@type='submit']")
    login_button.submit()
    sleep(1)
except Exception as e:
    error_flg = True
    print("ユーザー名、パスワード入力時にエラーが発生しました。詳細:", str(e))
```

```python
if error_flg is False:
    try:
        sleep(3)
        notnow_button = driver.find_element(By.XPATH,  "//div[@role='button' and text()='後で']")
        notnow_button.click()
        sleep(1)
    except Exception as e:
          print("「後で」ボタンをクリックする際にエラーが発生しました。詳細:", str(e))
```
